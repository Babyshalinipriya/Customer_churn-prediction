```python
def plot_importance(model, features, dataframe, save=False):
    # Check if the model has the 'feature_importances_' attribute
    if hasattr(model, 'feature_importances_'):
        num = len(dataframe)
        feature_imp = pd.DataFrame({"Value": model.feature_importances_, "Feature": features.columns})
        plt.figure(figsize=(10, 10))
        sns.set_theme(font_scale=1)
        sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value", ascending=False)[0:num])
        plt.title("Features")
        plt.tight_layout()
        plt.show()
        if save:
            plt.savefig("importances.png")
    else:
        print("The provided model does not have feature importances.")

# Assuming 'best_gb_model' is your Gradient Boosting model
plot_importance(best_gb_model, x_train, data)
```
